## Transforming Perspectives
### Using Computer Vision Tecniques
In this notebook we'll define a perspective transformation procedure for 'unwarping' images.

```python
# Import libraries
import numpy as np
import cv2
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import os

test_path = "./test_images/"
test_images = os.listdir(test_path)
```

```python
# plotting function for images
def plot(images, no_rows, no_cols, width = 10., height = 4., color='gray'):
    fig = plt.figure(1, (width, height))
    i = 0
    
    for image in images:
        sub = fig.add_subplot(no_rows, no_cols, i + 1)
        sub.imshow(image, color)
        sub.axis("off")
        i += 1

    plt.show()
```

```python
print("Image size: {}".format(mpimg.imread(os.path.join(test_path, test_images[0])).shape))
```

#### Define points of interest
Define the points of interest to perform the perspective transformation.

```python
src_map = np.float32([[684, 450],
                      [984, 640],
                      [316, 640],
                      [596, 450]])

dst_map = np.float32([[980, 0],
                      [980, 720],
                      [320, 720],
                      [320, 0]])
```

Show the points relative to the image.

```python
test_img = mpimg.imread(os.path.join(test_path, test_images[0]))
plt.imshow(test_img)
for x,y in src_map:
    plt.plot(x, y, ".")
```

```python
### Define the warping functions ###

# Warp an image using the specified source and destination coordinate mappings.
def warp(img, src, dst):
    img_size = (img.shape[1], img.shape[0])
    M = cv2.getPerspectiveTransform(src, dst)
    return cv2.warpPerspective(img, M, img_size, flags=cv2.INTER_LINEAR)

# Unwarps an image using the specified source and destination coordinate mappings.
def unwarp(img, src, dst):
    img_size = (img.shape[1], img.shape[0])
    Minv = cv2.getPerspectiveTransform(dst, src)
    return cv2.warpPerspective(img, M, img_size, flags=cv2.INTER_LINEAR)
```

```python
# plot the warped images side by side

sampled = []
for x in test_images:
    img = mpimg.imread(os.path.join(test_path, x))
    
    warped = warp(img, src_map, dst_map)
    sampled.append(img)
    sampled.append(warped)

plot(sampled, 4, 4, 12., 8.)
```
